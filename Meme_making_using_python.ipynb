```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

```python
img = Image.open("/content/I-guide-others-to-a-treasure-I-cannot-possess-meme-9.jpg")
print("Image Size:", img.size)
print("Image Mode:", img.mode)
plt.imshow(img)
```

```python
DRAW = ImageDraw.Draw(img)

# Preparing the Text Element
TEXT_1 = """HR with 5 years experience & 6 LPA salary
hiring fresher/collge passouts for 40LPA"""

# Include your own .ttf font (specify the correct path to the font file)
font_path = "/content/Poppins-Bold.ttf"
font = ImageFont.truetype(font_path, 22)

# Draw the text
DRAW.text((10, 30), text=TEXT_1, fill=(0, 0, 0), font=font, align="left")

# Display the image with matplotlib
plt.imshow(img)
plt.show()
```
